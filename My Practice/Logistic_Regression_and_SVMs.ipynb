```python
# StandardScaling, Training
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# X_test ----> Standard Scaled ---> transform ---> X_test_transformed ---> Logistic Regression ----> predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

## SVC

```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
data = load_breast_cancer()
```

```python
X = data['data'][:, :2]
y = data['target']
```

```python
def train_and_plot(kernel, gamma, C):
    svm = SVC(kernel=kernel, gamma=gamma, C=C)
    svm.fit(X, y)

    # Plot decision boundary
    DecisionBoundaryDisplay.from_estimator(
        svm,
        X,
        response_method="predict",
        xlabel=data['feature_names'][0],
        ylabel=data['feature_names'][1]
    )

    # Scatter plot
    plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolors='k')
    plt.show()
```

```python
train_and_plot('rbf', 0.5, 1)
```

```python
train_and_plot('rbf', 0.5, 100)
```

```python
train_and_plot('rbf', 0.5, 1)
```

```python
train_and_plot('rbf', 1, 1)
```

```python
train_and_plot('rbf', 0.1, 1)
```

```python
train_and_plot('linear', 0.5, 0.5)
```

```python
train_and_plot('poly', 0.5, 0.5)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001]
}

grid = GridSearchCV(SVC(), param_grid, verbose=3)

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
accuracy_score(y_test, grid_predictions)
```
