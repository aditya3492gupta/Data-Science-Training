```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data
```

```python
X = data['data']
```

```python
y = data['target']
```

```python
X.shape
```

```python
y.shape
```

```python
import numpy as np
```

```python
y_reshaped = np.reshape(y, (569, 1))
```

```python
dfdata = np.concatenate([X, y_reshaped], axis=1)
```

```python
dfdata.shape
```

```python
col_names = data['feature_names']
col_names = np.append(data['feature_names'], ['target'])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dfdata, columns=col_names)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
dfdata = np.concatenate([X_scaled, y_reshaped], axis=1)
```

```python
np.mean(X_scaled), np.std(X_scaled)
```

```python
normalized_df = pd.DataFrame(dfdata, columns=col_names)
```

```python
normalized_df
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
comp = pca.fit_transform(X_scaled)
```

```python
comp.shape
```

```python
compdfdata = np.concatenate([comp, y_reshaped], axis=1)
```

```python
comp_df = pd.DataFrame(compdfdata, columns=['PC1', 'PC2', 'target'])
```

```python
comp_df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 10))
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA (n=2)")
plt.scatter(comp_df['PC1'], comp_df['PC2'], c=comp_df['target'])
plt.legend([0, 1])
plt.show()
```

## Hyperparameter Optimization Revision

```python
import numpy as np
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1, 0.5, 0.75, 0.9, 1, 1.1, 1.25, 1.5, 10]}
```

```python
grid_search = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=5, verbose=3)
grid_search.fit(X, y)
```

```python
grid_search.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import norm
```

```python
param_dist={'C': norm(loc=2, scale=1)}
```

```python
random_search = RandomizedSearchCV(SVC(kernel='rbf'), param_dist, n_iter=10, verbose=3)
random_search.fit(X, y)
```

```python
random_search.best_params_
```
