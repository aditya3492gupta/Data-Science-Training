```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Dummy data
data = {
    'age': [25, 30, 35, None, 40],
    'income': [50000, 60000, None, 70000, 80000],
    'gender': ['M', 'F', 'F', 'M', 'M'],
    'education': ['Bachelor', 'Master', 'PhD', 'Master', 'Bachelor'],
    'target': [0, 1, 0, 1, 1]
}

df = pd.DataFrame(data)

df.head()
```

```python
X = df.drop('target', axis=1)
```

```python
X
```

```python
numeric_features = ['age', 'income']
categorical_features = ['gender', 'education']
```

```python
# Numeric Transformer
numeric_transformer = Pipeline(
    steps = [
        ('imputation', SimpleImputer(strategy='median')),
        ('scaling', StandardScaler())
    ]
)
```

```python
# Categorical Transfomer
categorical_transformer = Pipeline(
    steps = [
        ('imputation', SimpleImputer(strategy='most_frequent')),
        ('encoding', OneHotEncoder(handle_unknown='ignore'))
    ]
)
```

```python
transformer = ColumnTransformer(
    transformers = [
        ('numeric_transformations', numeric_transformer, numeric_features),
        ('categorical_transformations', categorical_transformer, categorical_features)
    ]
)
```

```python
X_processed = transformer.fit_transform(X)
```

```python
X_processed
```

```python
dir(transformer)
```

```python
df = pd.DataFrame(X_processed, columns = list(numeric_features) + list(transformer.named_transformers_['categorical_transformations']['encoding'].get_feature_names_out()))
```
