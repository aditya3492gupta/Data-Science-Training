```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('amazon_file.csv')
```

```python
df.dropna(subset=['rating_count'], inplace=True)
```

```python
df.describe(include='all')
```

```python
namechange={"product_id":"Product ID", "product_name":"Product Name", "category":"Category", "discounted_price":"Discounted Price", "actual_price":"Actual Price", "discount_percentage":"Discount Percentage", "rating":"Rating", "rating_count":"Rating Count", "about_product":"About Product", "user_id":"User ID", "user_name":"User Name", "review_id":"Review ID", "review_title":"Review Title", "review_content":"Review Content", "img_link":"IMG Link", "product_link":"Product Link"}
df = df.rename(columns=namechange)
```

```python
df
```

```python
df.drop('Product Link',axis=1,inplace=True)
df.drop('IMG Link',axis=1,inplace=True)
df 
```

```python
df.drop('Review ID',axis=1,inplace=True)
df
```

```python
df.duplicated().sum()
print("Dataset remains as ", df.shape[0], " rows and ", df.shape[1], " columns")
```

```python
df['Discounted Price'] = df['Discounted Price'].str.replace("₹",'').str.replace(",",'')
df['Actual Price'] = df['Actual Price'].str.replace("₹",'').str.replace(",",'')
df['Discount Percentage'] = df['Discount Percentage'].str.replace("%",'')
df['Rating Count'] = df['Rating Count'].str.replace(",",'')
```

```python
df.head()
```

```python
unique_ratings = df['Rating'].unique()
print(unique_ratings)
df = df.loc[df['Rating'] != '|']
```

```python
data = df.astype({'Discounted Price': 'float64', 'Actual Price': 'float64', 'Discount Percentage': 'float64', 'Rating': 'float64', 'Rating Count': 'float64'})
```

```python
category_counts = data['Product ID'].value_counts().sort_index()
category_counts
# Plotting
plt.figure(figsize=(10, 6))
category_counts.plot(kind='line', marker='o', color='blue')
plt.title('Number of Products in Each ID')
plt.xlabel('Product ID')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_sorted = data.sort_values(by='Rating', ascending=False)
df_sorted
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Rating'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
top_n = 5
top_rated_products = df_sorted.head(top_n)
plt.figure(figsize=(10, 6))
plt.plot(top_rated_products['Product ID'], top_rated_products['Rating'], color='skyblue')
plt.title('Rating Against Product ID')
plt.xlabel('Product ID')
plt.ylabel('Rating')
plt.ylim(0, 5)  # Set the y-axis limit to the range of ratings (0 to 5)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
